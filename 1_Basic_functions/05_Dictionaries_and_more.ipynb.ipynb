# Loops

```python
# there are two main types of loops WHILE and FOR
# firstly we will extract data from the list using FOR loop
numbers = [0,1,2,3,4,5]
```

```python
# FOR loops generates on iterable object
for number in numbers:
  print(number)
```

```python
# WHILE loops generate until the condition meets
count = 1
while count <= 10:
  print(count)
  count += 1
```

```python
# to print a string with given condition
job_openings = 6
job_index = 0
position_exp = [5,4,6,2,3,4]
while job_index < job_openings:
  print('Position requires',position_exp[job_index],'year of experience')
  job_index += 1
```

# List Comprehension

```python
# lets try to make a list first
numbers = [x for x in range(10)]
numbers
```

```python
# to create a position requirement list
position_exp_req = [1,2,3,4]
# now we want to print each instance
for x in position_exp_req:
  print('Position requires',x,'year of experience')
```

```python
# lets try a shorter one
exp = [x for x in position_exp_req]
exp
```

```python
user_exp = 2
user_exp
```

```python
# create a list of job positions for which the user is qualified

qualified_positions = [x for x in position_exp_req if user_exp >= x]
qualified_positions
```
