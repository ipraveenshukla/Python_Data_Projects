# Conditional Statements

```python
# so now we'll how conditional statements work
if True:
  print('Everything is fine')
```

```python
# now with some variables
skill = 'SQL'
exp = 10
skill_required = 'SQL'
result = 'Skill matched!'
```

```python
if skill == skill_required:
  print (result)
```

```python
if skill == skill_required and exp < 2:
  print('Junior Data Analyst')
elif  skill == skill_required and exp >= 2 and exp <= 5:
  print('Mid Data Analyst')
elif skill == skill_required and exp > 5:
  print('Senior Data Analyst')
```

```python
# now we have know how else and pass works
if skill == skill_required and exp < 2:
  print('Junior Data Analyst')
else:
  print('No match found!')
```

```python
# Now let's start using !=
# now with some variables
skill = 'SQL'
exp = 3
skill_required = 'SQL'
result = 'Skill matched!'

if skill != skill_required:
  print('No match found!')
```

```python
#now lets try and with == operator
if skill != skill_required:
  print ('No skill match')
elif skill == skill_required and exp < 2:
  print('More Experience Needed')
elif skill == skill_required and exp >= 2:
  print('Candidate Eligible')
```

# Types of Lists

```python
# here we will understand list and how to use them
# firstly we will assign the values to the variable
job_skills = ['Python', 'SQL','Excel']
```

```python
job_skills
```

```python
# now using indexing we will extract specific skill to output
job_skills[1]
```

```python
# Now using same indexing method we will replace the value of specific skill
job_skills[1] = 'Tableau'
```

```python
# print the updated list
job_skills
```

```python
# now we will add a skill to the list
job_skills.append('looker')
```

```python
job_skills
```

```python
# to know how many items are present on the list (We have 5 items)
len(job_skills)
```

```python
# now to insert a specific item to a specific place we have to provide index value and item name
job_skills.insert(0, 'Power BI')
```

```python
job_skills
```

```python
# now to remove a certain item we have to place the item name with remove function, then update the list
job_skills.remove('looker')
```

```python
job_skills
```

```python
# now i want to join two lists
job_skills_2 = ['Python', 'SQL','Excel']
# and
job_skills_3 = ['Power BI', 'Tableau', 'Looker']
job_skills_all = job_skills + job_skills_2 + job_skills_3
```

```python
# now to print all the cumulative skills
job_skills_all
```

```python
# now we will append items form one list to another
job_skills_2.extend(job_skills_3)
```

```python
job_skills_2
```

```python
# to concatenate
print('I have these skills: '+', '.join(job_skills_all))
```

```python
# to extract specific columns
job_skills_all[:1]
```

```python
# full list
job_skills_all[:]
```

```python
# to get last 3 items
job_skills_all[-3:]
```

```python
# to get every alternative items out of the list
job_skills_all[::3]
```

```python
# to do the unpacking
s1, s2, s3, s4, s5, s6, s7, s8, s9, s10 = job_skills_all
print(s1)
print(s2)
print(s3)
```

```python
# to unpack the list we use * operator
s1, s2, *job_skills_all = job_skills_all
print(s1)
print(s2)
print(job_skills_all)
```
