```python
# importing library
import pandas as pd
from datasets import load_dataset
```

```python
# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# now data clean up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
# now we want to find jobs for specific country
df_india = df[df['job_country'] == 'India']
df_india.head()
```

```python
# now we want to see the median salary for different job_title_short in india
df_india.groupby('job_title_short')['salary_year_avg'].mean()
```

```python
# suppose we want other stats as well
df_india.groupby('job_title_short')['salary_year_avg'].agg(['mean', 'median', 'min', 'max'])
```

```python
# now supoose we want to sort values by median in desc order
df_india.groupby('job_title_short')['salary_year_avg'].agg(['median', 'mean', 'min', 'max']).sort_values('median', ascending=False)
```
