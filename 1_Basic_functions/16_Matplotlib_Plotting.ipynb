```python
# Necessary steps
import pandas as pd
from datasets import load_dataset

# import matplotlib
import matplotlib.pyplot as plt

# load dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# finally data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# after carefully reading the documentation plot() function plot the point and we have to specify x and y axis as well
plt.plot(df['job_posted_date'], df['job_posted_date'])
plt.show()
# we can see the graph below
```

```python
# we can also compress the instructions code
plt.plot('job_posted_date', 'job_posted_date', data = df)
plt.show()
```

```python
# now we want to specify how many jobs were released during the given dates
date_counts = df['job_posted_date'].value_counts().sort_index()
date_counts
```

```python
# suppose i want to plot the given data on the plot; dates on x axis and no of jobs released on y axis
plt.plot(date_counts.index, date_counts.values)
plt.show()
```

```python
# as given graph is showing data on daily bases by i want it for monthly basis
df['job_posted_month'] = df['job_posted_date'].dt.month

# now we have inserted the new column job_posted_month, its time to group by
monthly_counts = df['job_posted_month'].value_counts().sort_index()
monthly_counts
```

```python
# now we can plot our graph
plt.plot(monthly_counts.index, monthly_counts.values)
plt.show()
```

```python
# so now we will look for total job count based on specific job_title_short
job_count = df['job_title_short'].value_counts().sort_values(ascending = True)
job_count

# we can also specify top or bottom attributes given below 🔽
# job_count = df['job_title_short'].value_counts().head(5)
```

```python
# now we want to plot the bar graph of the same
plt.barh(job_count.index, job_count.values)
plt.show()
```

```python
# further we can also give axis titles
plt.barh(job_count.index, job_count.values)

# Adding titles and labels
plt.title('What are Most Demanding Jobs?')
plt.xlabel('Job Count')
plt.ylabel('Title')

plt.show()
```

```python
# Now the Same graph in vertical
job_count_v = df['job_title_short'].value_counts().head(7)
plt.bar(job_count_v.index, job_count_v.values)

# Adding titles and labels
plt.title('What are Most Demanding Jobs?')
plt.xlabel('Job Count')
plt.ylabel('Title')

plt.xticks(rotation = 45, ha='right')

plt.show()
```
