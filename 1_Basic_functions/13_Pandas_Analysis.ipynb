```python
import pandas as pd
```

```python
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.describe()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
# i want to create a new column named month name
df['month_name'] = df['job_posted_date'].dt.month_name()
```

```python
# the number of jobs posted each month by month name in proper order (jan, feb, mar)
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
job_counts = df['month_name'].value_counts().reindex(months_order)
job_counts
```
