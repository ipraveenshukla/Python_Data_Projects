```python
# lets import libraries first
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Now Dataset 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# sort by date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
# firstly we have to define a function in which project inflated salary will be 5% higher
def inflation(x):
    return x * 1.05

# now lets apply this function to the salary_year_avg column
df['salary_year_inflated'] = df['salary_year_avg'].apply(inflation)
df.head(10)

# # now we have remove NaN values
# df['salary_year_inflated'] = df['salary_year_inflated'].notna()
# df.head(10)

# lets delete NaN from salary_year_avg column
# df = df.dropna(subset=['salary_year_avg'])
# df

# now lets get the data frame for job title short, salary_year_avg and salary_year_inflated
df_inflated = df[['job_title_short','salary_year_avg','salary_year_inflated']]
df_inflated.reset_index()
df_inflated.set_index('job_title_short')
```

```python
# now we want only filled salary_year_avg columns
df_inflated = df_inflated.dropna(subset=['salary_year_avg'])
df_inflated.sample(10)
```
