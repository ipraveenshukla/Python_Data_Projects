```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Import Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Now Sort by date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# we are getting data for india only
df_india = df[df['job_country'] == 'India'].copy()
id(df_india) == id(df)
```

```python
# lets extract month from each job posted date
df_india['month_name'] = df_india['job_posted_date'].dt.strftime('%B')
df_india
```

```python
# now lets make pivot table based on month
df_india_p = df_india.pivot_table(index='month_name',columns='job_title_short',aggfunc='size')
df_india_p
```

```python
# now lets format the pivot table accordingly
df_india_p = df_india_p.reset_index()
df_india_p['month_number'] = pd.to_datetime(df_india_p['month_name'],format='%B').dt.month
df_india_p = df_india_p.sort_values('month_number')
df_india_p = df_india_p.set_index('month_name')
df_india_p = df_india_p.drop(columns='month_number')
df_india_p
```

```python
# now finally lets print the chart
df_india_p.plot(kind='line')
```

```python
# we want top 3 jobs so...
top_3 = df_india['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
top_3
```

```python
# now with the top 3 list 
df_india_p[top_3].plot(kind='line')

# now final formatting 
plt.xlabel('Month')
plt.ylabel('Job Count')
plt.title('What are Most Demanding Jobs in India?')
plt.legend()
plt.show()
```
