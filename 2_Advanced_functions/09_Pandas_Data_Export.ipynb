# Lets Export the Data to different formats

```python
# Import library
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# sort by date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# for example we want to create a copy of dataframe in which job_country is United States
df_us = df[df['job_country'] == 'United States'].copy()

# sort this dataframe by date
df_us.sort_values(by='job_posted_date', inplace=True)
df_us
```

```python
# now lets export this in different formats
# csv
df_us.to_csv('data_jobs_us.csv',index=False)
```
