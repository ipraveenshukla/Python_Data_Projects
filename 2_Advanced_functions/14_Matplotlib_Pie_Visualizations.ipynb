# Pie Charts Combined

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Loading Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Sorting Dataframe
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# converting job_skills to list
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
```

```python
# creating new dataframe for Data Scientist
df_DS = df[df['job_title_short']== 'Data Scientist'].copy()
```

```python
# list for work from home jobs
work_from_home = df['job_work_from_home'].value_counts()
degree_mention = df['job_no_degree_mention'].value_counts()
health_insurance = df['job_health_insurance'].value_counts()
```

```python
# plotting the pie chart
plt.pie(work_from_home,labels=[False, True],startangle=90,autopct='%1.1f%%')
plt.title('WFH status')
plt.show()
```

```python
# creating subplots for other columns (job_work_from_home, job_no_degree_mention and job_health_insurance)
fig, ax = plt.subplots(1,3)
fig.set_size_inches((13,5))

plt.subplot(1,3,1)
plt.pie(work_from_home,labels=[False, True],startangle=90,autopct='%1.1f%%')
plt.title('Is job Work From Home?')

plt.subplot(1,3,2)
plt.pie(degree_mention,labels=[False, True],startangle=90,autopct='%1.1f%%')
plt.title('Is job Without Degree?')

plt.subplot(1,3,3)
plt.pie(health_insurance,labels=[False, True],startangle=90,autopct='%1.1f%%')
plt.title('Is Health Insurance Provided?')

plt.show()
```
