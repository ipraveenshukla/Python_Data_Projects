```python
# firstly we have to import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# now lets import dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# now sort by date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# now we will create copy using the copy function
df_altered = df.copy()
df_altered.info()
```

```python
# now we have to do replace Nan with '0'
df_altered.loc[:5,['salary_rate','salary_hour_avg','salary_year_avg']]
```

```python
# ;lets get median salary first
median_salary = df_altered['salary_year_avg'].median()
median_salary
```

```python
# now we will fill all NaN values with median salary
df_altered['salary_year_avg'] = df_altered.loc[:,'salary_year_avg'].fillna(median_salary)

# now lets print it
df_altered.head(10)
```
