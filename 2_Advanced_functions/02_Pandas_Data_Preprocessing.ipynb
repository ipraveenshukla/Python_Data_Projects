```python
# importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# loading datset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# sorting by date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df_cleaned = df.dropna(how='all')
```

```python
df_cleaned.info()
```

```python
# looking for specific columns
df_cleaned.iloc[:9,11:14]
```

```python
# now we will fill those values with '0'
df_fill = df_cleaned.fillna(0)

# now seeing the result
df_fill.iloc[:9,11:14]
```

```python
# now we have to delete duplicate rows
df_unique = df_fill.drop_duplicates()

# checking the info
df_unique.info()

# we can see rows value is reduced from 785741 to 785640 
reduced_rows = 785741 - 785640
reduced_rows
```

```python
# now if we try to remove certain job title because sometime same job is posted more than once
df_job_title_unique = df_fill.drop_duplicates(subset='job_title')

# now lets check unique job titles
df_job_title_unique.info()
```
