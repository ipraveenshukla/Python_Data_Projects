```python
# firstly we will load libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# now we will load the datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# now data clean up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# now we will use iloc function to locate certain columns
df.iloc[0]
# to get top 10 rows
df.iloc[0:10]
```

```python
# Now we will locate certain items in table
# df.info()
df.iloc[[56414],[0,10,13]]
```

```python
# getting the first 10 rows of dataframe
df.iloc[:9]
```

```python
# to get first 5 columns and 5 rows of the dataframe
df.iloc[:6,:5]
```

```python
# by loc functiom we can specify more 
df.loc[:10,['job_title_short','job_country']]
```

```python
# similarily we can also provide range for that
df.loc[100:120,'job_posted_date':'job_country']
```
